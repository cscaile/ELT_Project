```python
# Dependencies
from bs4 import BeautifulSoup
import requests
```

```python
url= "https://www.sailboatlistings.com/location/Colorado"
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object
soup = BeautifulSoup(response.text)
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# results are returned as an iterable list
tr = soup.find_all('tr')
print(tr)
```

```python
# Loop through returned results
for result in tr:
    # Error handling
    try:
        
        length=result.find('td', background="/tb/b1.jpg")
        year=result.find('td', background="/tb/b2.jpg")
        make=result.find('td', background="/tb/b3.jpg")
        model=result.find('td', background="/tb/b4.jpg")
        city=result.find('td', background="/tb/b5.jpg")
        state=result.find('td', background="/tb/b6.jpg")
        price=result.find('td', background="/tb/b7.jpg")

        # Print results only if title, price, and link are available
        if (length and year):
            print('-------------')
            print(length.text)
            print(year.text)
            print(make.text)
            print(model.text)
            print(city.text)
            print(state.text)
            print(price.text)
           # print(link)
    except AttributeError as e:
        print(e)
```

```python
#Turn results into a list of dictionaries to create a pandas dataframe with

sailboats=[]


for result in tr:
    
    try:
        data={}
        
        length=result.find('td', background="/tb/b1.jpg")
        year=result.find('td', background="/tb/b2.jpg")
        make=result.find('td', background="/tb/b3.jpg")
        model=result.find('td', background="/tb/b4.jpg")
        city=result.find('td', background="/tb/b5.jpg")
        state=result.find('td', background="/tb/b6.jpg")
        price=result.find('td', background="/tb/b7.jpg")

        # Print results only if title, price, and link are available
        if (length and year):
            
            data['length']=(length.text)
            data['year']=(year.text)
            data['make']=(make.text)
            data['model']=(model.text)
            data['city']=(city.text)
            data['state']=(state.text)
            data['price']=(price.text)
            sailboats.append(data)
           
    except AttributeError as e:
        print(e)
        
        
        
print(sailboats)
```

```python
import pandas as pd

sailboats_df=pd.DataFrame(sailboats)
sailboats_df.drop([0])
```

```python
#Turn missed results into another dataframe

sailboats2_df=pd.DataFrame(sailboats_2)
sailboats2_df
```

```python
#Merge the 2 dataframes for 1 complete dataframe

complete_sailboats=sailboats_df.append(sailboats2_df)
complete_sailboats=complete_sailboats.reset_index(drop=True)
complete_sailboats=complete_sailboats.drop([0])
complete_sailboats
```
