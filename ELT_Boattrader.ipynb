```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# URL of page to be scraped
url = 'https://www.boattrader.com/boats/'
```

```python
# Retrieve page with the requests module
browser.visit(url)
```

```python
# Iterate through all pages

for x in range(50):
    # HTML object
    page = requests.get(f'https://www.boattrader.com/boats/page-{x}')
    # Parse HTML with Beautiful Soup
    soup = BeautifulSoup(page.text, 'html.parser')
    # Retrieve all elements that contain book information
    headers = soup.find_all('li', {'data-reporting-click-listing-type': 'standard listing'})
      # Iterate through each book
    for header in headers:
        try:
            # Identify and return title of listing
            title = header.find('div', class_="name").text
            # Identify and return price of listing
            price =header.find('div', class_="price").text
            # Identify and return link to listing
            link = header.a['href']
            

            # Print results only if title, price, and link are available
            if (title and price and link):
                print('-------------')
                print(title)
                print(price)
                print(link)
            
        # this prevents overloading the website    
#             time.sleep(1) 
        except AttributeError as e:
            print(e)
            
print("Scraping Complete")    
```
