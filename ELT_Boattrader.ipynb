```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# URL of page to be scraped
url = 'https://www.boattrader.com/boats/'
zip = 80202
radius = 100
```

```python
# Retrieve page with the requests module
browser.visit(url)
```

```python
# Create a list of column headers and data dictionary
columns = ['Boat', 'Price', 'Website', 'Source']
data = ({'Boat':[],
         'Price':[],
         'Website':[],
         'Source':[]
        })

# Designating data source
website = 'Boat Trader'
# Creating an open list for each boats' listing site
site = []

# Iterate through all pages
for x in range(50):
    # HTML object
    page = requests.get(f'https://www.boattrader.com/boats/zip-{zip}/radius-{radius}/page-{x}')
    # Parse HTML with Beautiful Soup
    soup = BeautifulSoup(page.text, 'html.parser')
    # Retrieve all elements that contain boat information
    headers = soup.find_all('li', {'data-reporting-click-listing-type': 'standard listing'})
    
    # Iterate through each boat
    for header in headers:
        try:
            # Identify and return title of listing
            boat_name = header.find('div', class_="name").text
            data['Boat'].append(boat_name)
            # Identify and return price of listing
            boat_price = header.find('div', class_="price").text
            data['Price'].append(boat_price)
            # Identify and return link to listing
            web_link = header.a['href']
            data['Website'].append(web_link)
            site.append(web_link)
            data['Source'].append(website)
            
            # Print results only if title, price, and link are available
            if (title and price and link):
                print('-------------')
                print(boat_name)
                print(boat_price)
                print(web_link)
            
        # this prevents overloading the website    
#             time.sleep(1) 
        except AttributeError as e:
            print(e)
            
print("Scraping Complete")    
```

```python
# Create a list of column headers and data dictionary
info_col = ['Year', 'Make', 'Model', 'Class', 'Length']
info = ({'Year':[],
         'Make':[],
         'Model':[],
         'Class':[],
         'Length':[]
        })

# Iterate through the boat websites to get more information
for site in site:
    page = requests.get(site)
    # Parse HTML with Beautiful Soup
    soup = BeautifulSoup(page.text, 'html.parser')
    
    # Open list for boat specifications
    boat_spec=[]
    
    # Iterate through all datatable-value items in the html to pull boat specs
    for item in soup.select("div.datatable-value"):
        boat_spec.append(item.get_text())

    # Append boat specs to dictionary
    info['Year'].append(boat_spec[0])
    info['Make'].append(boat_spec[1])
    info['Model'].append(boat_spec[2])
    info['Class'].append(boat_spec[3])
    info['Length'].append(boat_spec[4])
```
