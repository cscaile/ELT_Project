```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# URL of page to be scraped
url = 'https://denver.craigslist.org/search/sss?query=boats&sort=rel'
```

```python
# Retrieve page with the requests module
browser.visit(url)
```

```python
# Iterate through all pages
for x in range(50):
    # HTML object
    html = browser.html
    # Parse HTML with Beautiful Soup
    soup = BeautifulSoup(html, 'html.parser')
    # Retrieve all elements that contain book information
    articles = soup.find_all('li', class_="result-row")
      # Iterate through each book
    for article in articles:
        try:
            # Identify and return title of listing
            title = article.find('a', class_="result-title").text
            # Identify and return price of listing
            price = article.a.span.text
            # Identify and return link to listing
            link = article.a['href']
            

            # Print results only if title, price, and link are available
            if (title and price and link):
                print('-------------')
                print(title)
                print(price)
                print(link)
            
        # this prevents overloading the website    
#         time.sleep(1) 
        except AttributeError as e:
            print(e)
            
    
# Click the 'Next' button on each page
    try:
        browser.click_link_by_partial_text('next >') 
          
    except:
        print("Scraping Complete")    
        break
```
